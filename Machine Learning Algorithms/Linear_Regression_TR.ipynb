**LİNEER REGRESYON**
*   Bilinen bir veri değeri kullanarak, bilinmeyen bir veri değerini bulma işlemine lineer regresyon denir.
*   Bilinen veriler ile bilinmeyen veriler arasında doğrusal bir denklem kullanılarak bilinmeyen değer bulunur.
*   İki değişken arasında çizgi grafiği çizmeyi hedefler.
*   Bilinen değişken(bağımsız değişken, x) ile, bilinmeyen değişken(bağımlı değişken, y) değerini bulmak için **y = c*x + b** formülü kullanılır.



> Lineer Regresyon, iki değişken arasında doğrusal ilişki olduğu durumlarda kullanılır.

*   Lineer regresyonda cost(loss) fonksiyonu ortalama kareler hatası kullanılarak bulunur. **(Mean Square Error MSE)**

image.png


*   MSE yerine ortalama mutlak değer **(Mean Absolute Error)** kullanılır.

**IMPORT İŞLEMLERİ**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

**DATASET YÜKLEME**

```python
dataset = pd.read_csv('https://github.com/ybifoundation/Dataset/raw/main/Salary%20Data.csv')
```

```python
dataset.columns
```

```python
X = dataset[['Experience Years']]
Y = dataset[['Salary']]
```

**SCATTER PLOT İLE VERİLERİ GÖRSELLEŞTİRME**

```python
plt.scatter(X,Y)
```

**LİNEER REGRESYON MODEL EĞİTİM**

```python
model = LinearRegression().fit(X, Y)
```

```python
print(model.intercept_[0])
print(model.coef_[0][0]) #eğim katsayısı
```

**MODEL TAHMİN İŞLEMLERİ**

```python
Y_pred = model.predict(X)
mae = mean_absolute_error(Y, Y_pred)
mse = mean_squared_error(Y, Y_pred)
#hata değerleri; modelin deneyim yılına göre ücret tahminini ne kadar iyi yaptığı
#düşük olması modelin iyi çalıştığını gösterir
print(mae)
print(mse)
```

```python
plt.plot(Y,color='r',label='Salary Value')
plt.plot(Y_pred,color='b',label='Predicted Salary Value')
plt.xlabel('Experience Years')
plt.ylabel('Salary')
plt.legend()
```
